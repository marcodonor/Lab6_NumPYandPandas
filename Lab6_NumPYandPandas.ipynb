```python
import numpy as np
import pandas as pd

# Step 1: Load dataset (make sure the path is correct!)
df_kaggle = pd.read_csv("survey_results_public.csv")

# Step 2: Select relevant columns
df_subset = df_kaggle[['Country', 'EdLevel', 'YearsCodePro', 'ConvertedComp']]

# Step 3: Clean the data
df_clean = df_subset.dropna()

# Step 4: Categorize experience into groups
df_clean['YearsCodePro'] = pd.to_numeric(df_clean['YearsCodePro'], errors='coerce')
df_clean = df_clean.dropna(subset=['YearsCodePro'])
df_clean['ExperienceLevel'] = np.where(df_clean['YearsCodePro'] >= 10, 'Senior', 'Junior')

# Step 5a: Group by **EdLevel** instead of Country
grouped_edlevel = df_clean.groupby('EdLevel')['ConvertedComp'].agg(['mean','median','std']).reset_index()
print("Grouped by Education Level:\n", grouped_edlevel.head())

# Step 5b: Group by Country and compute median + std
grouped_country = df_clean.groupby('Country')['ConvertedComp'].agg(['mean','median','std']).reset_index()
print("Grouped by Country:\n", grouped_country.head())

# Step 6: Explore top 10 countries with highest average compensation
top10_countries = grouped_country.sort_values(by='mean', ascending=False).head(10)
print("Top 10 Countries by Average Compensation:\n", top10_countries)
```
